Time Series Analysis of  Monthly Sales Revenue [Line chart]

```python
# converting  order_date to Date_time
df['order_date'] = pd.to_datetime(df['order_date'])

# Extract year and month  for grouping
df['year_month'] =df['order_date'].dt.to_period('M')

# aggregate revenue by year_month
montly_sales = df.groupby('year_month')['Revenue'].sum().reset_index()

# converting the year_month back to string for plotting
montly_sales['year_month'] = montly_sales['year_month'].astype(str)

# creating the line chart

plt.Figure(figsize=(15,4))
sns.lineplot(data=montly_sales, x='year_month',y='Revenue',marker='o')
plt.title('Monthly Sales Revenue ', fontsize= 11)
plt.xlabel('Month', fontsize = 11)
plt.ylabel('Revenue ($)', fontsize =11)
plt.xticks(rotation = 45)
plt.grid(True)
plt.tight_layout()


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

# Convert 'order_date' to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Extract first day of month for consistent plotting
df['year_month'] = df['order_date'].dt.to_period('M').dt.to_timestamp()

# Aggregate revenue
monthly_sales = df.groupby('year_month')['Revenue'].sum().reset_index()

# Plot
plt.figure(figsize=(14, 4))
plt.plot(monthly_sales['year_month'], monthly_sales['Revenue'], color='navy', marker='o')

# Format x-axis as years with smart spacing
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_minor_locator(mdates.MonthLocator())
plt.gcf().autofmt_xdate()  

# Titles and labels
plt.title('Monthly Sales Trend', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue (Millions)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Filter out Jan and Feb 2018 if not already done
df_new = df[~((df['order_date'].dt.year == 2018) & (df['order_date'].dt.month.isin([1, 2])))].copy()

# Extract month name (abbreviated, e.g., Jan, Feb)
df_new['month'] = df_new['order_date'].dt.strftime('%b')

# Aggregate revenue by month across all years
monthly_trend = df_new.groupby('month')['Revenue'].sum().reset_index()

# Define correct calendar order
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Apply ordering to the month column
monthly_trend['month'] = pd.Categorical(monthly_trend['month'], 
                                        categories=month_order, 
                                        ordered=True)
monthly_trend = monthly_trend.sort_values('month')

# Plotting
plt.figure(figsize=(14, 4))
sns.lineplot(data=monthly_trend, x='month', y='Revenue', marker='o')
plt.title('Overall Monthly Sales Trend', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Bar Graph Of Top 10 And Bottom 10

```python
# Top 10 products by revenue
product_revenue = df_new.groupby('product_name')['Revenue'].sum().reset_index()
top_10_products = product_revenue.sort_values(by='Revenue', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
barplot = sns.barplot(data=top_10_products, x='product_name', y='Revenue', palette='tab10')

# Add data labels on top
for i, bar in enumerate(barplot.patches):
    height = bar.get_height()
    barplot.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:,.0f}',   
        ha='center',
        va='bottom',
        fontsize=7,
        fontweight='bold'
    )

plt.title('Top 10 Products by Revenue', fontsize=14)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group and get bottom 10 products by revenue
product_revenue = df_new.groupby('product_name')['Revenue'].sum().reset_index()
bottom_10_products = product_revenue.sort_values(by='Revenue', ascending=True).head(10)

# Data
x = bottom_10_products['product_name']
y = bottom_10_products['Revenue']

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(x, y, width= 0.6, color=sns.color_palette('pastel', n_colors=10)) 

# Add data labels
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:,.0f}',
        ha='center',
        va='bottom',
        fontsize=7,
        fontweight='bold'
    )

# Labels and formatting
plt.title('Bottom 10 Products by Revenue', fontsize=14)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
